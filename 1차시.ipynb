코들은 jupyter notebook이라는 웹 브라우저 통합 개발 환경을 제공하고 있다.   
아래의 설명과 함께 코들의 개발 환경을 사용해보자.
  

1. **코들 스튜디오에 입장한다.**

2. **새로운 노트북을 만든다.**  

새로운 노트북을 만드는 방법은 두 가지이다.
> ① [파일]→[새로 만들기]→[노트북]  
② [런처]→[노트북]→[python3(ipykernel)]  
좌상단의 + 버튼을 누르면 새로운 런처가 열립니다.    

  

3. **[+코드셀] 버튼을 클릭하면 파이썬 코드를 작성할 수 있는 셀이 생성된다. 생성된 셀에 파이썬 코드를 작성할 수 있다.**  
  

4. **작성한 코드를 실행해보자.**  

실행 방법은 두 가지이다.  
> ① 상단의 [실행] 버튼을 눌러 실행시킬 수 있다.   
② [ctrl + enter] 단축키로 실행시킬 수 있다.
