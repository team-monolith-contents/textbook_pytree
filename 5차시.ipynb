<div style="padding: 15px; border: 5px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #3c763d; border-color: #d6e9c6; font-weight: bold;">
<h2 style='font-weight: bold'>4. 예제로 while문 살펴보기</h2>

</div> 

```python
# 조건을 만족하지 않는 경우 반복문 종료
i = 0
while i<5:
    print(i)
    i += 1
```

```python
# 조건을 계속 만족해서 무한반복
i = 0

while i<5:
    print(i)
```

위와 같이코드를 작성하면 '0'이 무한 반복하여 출력됩니다.  
이렇게 무한 반복으로 프로그램이 종료되지 않는 경우에는 좌측 상단의 '중지' 버튼을 눌러 강제 종료하면 됩니다.   
출력된 결과를 보지 않으려면 셀에 커서를 둔 채로 우클릭을 하고, '셀 출력 지우기' 버튼을 눌러주면 됩니다.

```python
i = 0
while i < 5:
    print(i)
    i += 1

print('반복문 종료 후 변수 i에 저장된 값 : ', i)
```

<div style="padding: 15px; border: 5px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #3c763d; border-color: #d6e9c6; font-weight: bold;">
<h2 style='font-weight: bold'>5. 반복을 중단하는 break</h2>

</div> 

```python
while True:
    str_in = input('input : ')
    if str_in == 'exit':
        break
    print('입력문 : '+ str_in)
```

<div style="padding: 15px; border: 5px solid transparent; border-color: transparent; margin-bottom: 20px; border-radius: 4px; color: #3c763d; border-color: #d6e9c6; font-weight: bold;">
<h2 style='font-weight: bold'>6. 다음 반복으로 넘어가는 continue</h2>

</div> 

```python
n = 0
while n < 50:
    n += 1
    if n%2==0 or n%5==0:
        continue
    print(n, end=' ')

print('\n반복문 종료 후 변수 n에 저장된 값', n)
```
