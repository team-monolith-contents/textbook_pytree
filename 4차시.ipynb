```python
rain = int(input("비가 오면 1, 오지 않으면 0을 입력하세요. > "))

if rain :
    print("비가 오면 우산을 챙겨요")
```

```python
green_light = int(input("초록북: 1, 빨간불: 0을 입력하세요. > "))

if green_light==1:
    print("횡단보도를 건넌다.")
else:
    print("횡단보도를 기다린다.")
```

```python
season = input("계절을 입력하세요(summer, winter, or others) : ")

if season == 'summer':
    print('여름이면 바다로 가요 ', end='')
elif season == 'winter':
    print('겨울이면 스키장으로 가요  ', end='')
else:
    print('여름도 겨울도 아니면 산으로 가요 ', end='')
```
