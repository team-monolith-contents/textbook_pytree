```python
# 정답 확인
n = '파이썬'
p = '화이팅'
print(n+p)
print(n*3)
```

```python
n = '100'
a = '파이썬'
print(n+p)
```

```python
name = input("당신의 이름은?")
nick = input("당신의 별명은?")
print(name, '의 별명은', nick, '입니다.')
```

```python
num1 = int(input("첫 번째 숫자를 입력하세요: "))
num2 = int(input("두 번째 숫자를 입력하세요: "))
print(num2-num1)
```
