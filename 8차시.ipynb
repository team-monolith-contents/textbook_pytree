<h2 style='font-weight: bold'>5. 함수에 입력된 값이나 결과값을 나타내는 방법 알아보기</h2>

</div> 

```python
# 정답 확인
def A(x) :
    return x+2

def B(y) :
    y = y + 1
    return A(y*2)

def C(z) :
    return B(z+4)

print(C(5))
```
