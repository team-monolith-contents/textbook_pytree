<h2 style='font-weight: bold'>5. 키와 값으로 짝지은 딕셔너리</h2>

</div> 

```python
dct = {'apple':'사과', 'banana':'바나나', 'peach':'복숭아'}
```

**1. 키로 딕셔너리 값 참조하기**

```python
print(dct['apple'])
```

**2. 딕셔너리 값 참조하기**

```python
print(len(dct))
```

**3. 딕셔너리 키의 유무 확인하기**

```python
print('banana' in dct)
print('orange' in dct)
```

**4. 딕셔너리의 키에 해당하는 값 구하기**  
get()을 사용하여 딕셔너리에서 키에 해당하는 값을 구합니다.

```python
print(dct.get('peach'))
```

**5. 딕셔너리의 값 변경하기**

```python
dct = {'apple':'애플', 'banana':'바나나', 'peach':'복숭아'}
dct['apple'] = '사과'
print(dct)
```

> dct['apple'] = '사과'는 'apple' 키가 없는 경우 추가하고 있는 경우에는 값을 변경해요.

**6. 딕셔너리 키 구하기**

```python
dct = {'apple':'사과', 'banana':'바나나', 'peach':'복숭아'}
print(dct.keys())
```

**7. 딕셔너리 값 구하기**

```python
dct = {'apple':'사과', 'banana':'바나나', 'peach':'복숭아'}
print(dct.values())
```

**8. 딕셔너리의 항목을 값과 키로 구하기**

```python
dct = {'apple':'사과', 'banana':'바나나', 'peach':'복숭아'}
print(dct.items())
```

**1. 키와 값으로 만들기**  
키와 값으로 dict()을 사용하여 딕셔너리를 만들 수 있습니다. 이때, 키는 문자열이더라도 따옴표로 표시하지 않아요.

```python
dct = {'apple':'사과', 'banana':'바나나', 'peach':'복숭아'}
print(dct)
```

**2. 키 리스트와 값 리스트를 묶어서 만들기**  
자료형을 묶어주는 zip()을 사용하여 키와 값의 리스트를 묶어서 딕셔너리를 만들 수 있습니다.

```python
key = ['apple', 'banana', 'peach']
value = ['사과', '바나나', '복숭아']
dct = dict(zip(key, value))
print(dct)
```

**3. 튜플과 리스트로 만들기**  
데이터타입이 튜플 형태로 이루어진 리스트를 이용하여 딕셔너리를 만들 수 있습니다.  
이때 튜플은 (항목, 항목) 형태로 이루어져 있어야 합니다.  


```python
dct = dict([('apple', '사과'), ('banana', '바나나')])
print(dct)
```

**1. 항목 추가하기**

```python
dct = {'apple':'사과', 'banana':'바나나'}
dct.setdefault('peach', '복숭아')
print(dct)
```

**2. 항목 삭제하기**  
pop(키)로 키의 항목 값을 가져오고 딕셔너리에서는 삭제합니다.   
이때, 딕셔너리에는 키로 값을 가져오기 때문에 리스트처럼 pop()로 키를 생략해서 사용할 수 없습니다. 

```python
d = {'apple':'사과', 'banana':'바나나', 'peach':'복숭아'}
print(d.pop('apple'))
print(d)
```
